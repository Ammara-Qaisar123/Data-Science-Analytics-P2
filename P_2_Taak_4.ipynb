# Logistic Regression

Logistic Regression is one of the most commonly used algorithms for binary classification.

It predicts the probability that a patient has heart disease.

```python
# Import Required Libraries

from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier

from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
from sklearn.metrics import ConfusionMatrixDisplay
from sklearn.metrics import roc_auc_score
from sklearn.metrics import roc_curve

import matplotlib.pyplot as plt
```

```python
print(X_train.shape)
print(y_train.shape)
```

```python
print(type(X_train))
print(X_train.dtype)
```

```python
dt = DecisionTreeClassifier(random_state=42)

dt.fit(X_train,y_train)
```

```python
dt_pred = dt.predict(X_test)

dt_prob = dt.predict_proba(X_test)[:,1]
```

```python
dt_pred = dt.predict(X_test)

dt_prob = dt.predict_proba(X_test)[:,1]
```

```python
print("Accuracy : {:.2f}%".format(
    accuracy_score(y_test,dt_pred)*100
))
```

```python
print(classification_report(y_test,dt_pred))
```

```python
cm = confusion_matrix(y_test,dt_pred)

ConfusionMatrixDisplay(cm).plot(cmap='Greens')

plt.title("Decision Tree Confusion Matrix")

plt.show()
```

```python
dt_auc = roc_auc_score(y_test,dt_prob)

print("ROC AUC Score :",round(dt_auc,3))
```

```python
fpr,tpr,_ = roc_curve(y_test,dt_prob)

plt.figure(figsize=(6,5))

plt.plot(fpr,tpr,label="Decision Tree")

plt.plot([0,1],[0,1],'r--')

plt.xlabel("False Positive Rate")

plt.ylabel("True Positive Rate")

plt.title("ROC Curve")

plt.legend()

plt.show()
```

# Random Forest Classifier

Random Forest combines multiple decision trees to improve prediction accuracy and reduce overfitting.

```python
rf = RandomForestClassifier(
    n_estimators=200,
    random_state=42
)

rf.fit(X_train,y_train)
```

```python
rf_pred = rf.predict(X_test)

rf_prob = rf.predict_proba(X_test)[:,1]
```

```python
print("Accuracy : {:.2f}%".format(
    accuracy_score(y_test,rf_pred)*100
))
```

```python
print(classification_report(y_test,rf_pred))
```

```python
cm = confusion_matrix(y_test,rf_pred)

ConfusionMatrixDisplay(cm).plot(cmap='Oranges')

plt.title("Random Forest Confusion Matrix")

plt.show()
```

```python
rf_auc = roc_auc_score(y_test,rf_prob)

print("ROC AUC Score :",round(rf_auc,3))
```

```python
fpr,tpr,_ = roc_curve(y_test,rf_prob)

plt.figure(figsize=(6,5))

plt.plot(fpr,tpr,label="Random Forest")

plt.plot([0,1],[0,1],'r--')

plt.xlabel("False Positive Rate")

plt.ylabel("True Positive Rate")

plt.title("ROC Curve")

plt.legend()

plt.show()
```
